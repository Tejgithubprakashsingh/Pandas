```python
def grade_cal(score):
    if 300>score>240:
        return "A"
    elif 240>score>230:
        return "B"
    elif 230>score>220:
        return "C"
    elif score<220:
        return "Fail"
    else:
        return "Enter correct value"
```

```python
grade_cal(310)
```

```python
df['Grades']=df['Total'].apply(grade_cal)
df
```
